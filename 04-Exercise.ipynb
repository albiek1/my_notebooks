````python
import numpy as np
import matplotlib.pyplot as plt
filename = './data/befkbhalderstatkode.csv'
bef_stats_df = np.genfromtxt(filename, delimiter=',', dtype=np.uint, skip_header=1)
neighb = {1: 'Indre By', 2: 'Østerbro', 3: 'Nørrebro', 4: 'Vesterbro/Kgs. Enghave', 5: 'Valby', 6: 'Vanløse',
          7: 'Brønshøj-Husum', 8: 'Bispebjerg', 9: 'Amager Øst', 10: 'Amager Vest', 99: 'Udenfor'}
plotDict = {}
dd = bef_stats_df
mask = (dd[:,0] == 2015) & (dd[:,3] == 5100)
def n_pop_neighb(mask, n):
    pop = dd[mask & (dd[:,1] == n)]
    sum_pop = pop[:,4].sum()
    return sum_pop

for x, y in neighb.items():
    plotDict[y] = n_pop_neighb(mask, x)
    #print(y)
    #print(n_pop_neighb(mask, x))
    
plotDict_sorted = dict(sorted(plotDict.items(), key=lambda item: item[1], reverse=True))
plt.bar(plotDict_sorted.keys(), plotDict_sorted.values(), width=0.5, align='center')
plt.xticks(rotation=45, horizontalalignment='right',fontweight='light')
````

````python
age_mask = (dd[:,0]==2015)&(dd[:,2]>65)&(dd[:,1]<99)
print(dd[age_mask][:,4].sum())
````

````python
nordic_countries = {5110: 'Norge', 5120: 'Sverige', 5104: 'Finland', 5106: 'Island', 5101: 'Grønland'}
old_nords = 0
for x, y in nordic_countries.items():
    print(y)
    n_mask = (dd[:,0]==2015)&(dd[:,2]>65)&(dd[:,3]==x)
    country_sum = (dd[n_mask][0:,4].sum())
    print(country_sum)
    old_nords += country_sum
    
print('Result: ', old_nords)
````

````python
østerbro_pop = {}
vesterbro_pop = {}
for i in range(1992, 2016):
    ø_mask = (dd[:,1]==2)&(dd[:,0]==i)
    østerbro_pop[i] = (dd[ø_mask][:,4].sum())
    v_mask = (dd[:,1]==4)&(dd[:,0]==i)
    vesterbro_pop[i] = (dd[v_mask][:,4].sum())
    
plt.title("Population of Østerbro", fontsize=24)
plt.xlabel("year", fontsize=14)
plt.ylabel("popluation", fontsize=14)
plt.plot(østerbro_pop.keys(), østerbro_pop.values(), linewidth=5)
````
